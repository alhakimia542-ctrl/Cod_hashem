```python
import PyPDF2

# 1. فتح الملف الأصلي للقراءة
with open('US_Declaration.pdf', 'rb') as f:
    pdf_reader = PyPDF2.PdfReader(f)
    pdf_writer = PyPDF2.PdfWriter()

    # 2. نسخ الصفحات من القارئ إلى الكاتب
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        pdf_writer.add_page(page)

    # 3. حفظ النتيجة في ملف جديد (لأن الـ PDF لا يعدل في مكانه بسهولة)
    with open('Updated_Declaration.pdf', 'wb') as output_file:
        pdf_writer.write(output_file)
```

```python
import PyPDF2

f = open('US_Declaration.pdf', 'rb')

# القائمة التي ستحتوي على نصوص الصفحات
pdf_text = [0]  # وضعنا 0 ليكون نص الصفحة 1 في الفهرس 1

# استخدام الفئة الحديثة PdfReader
pdf_reader = PyPDF2.PdfReader(f)

# استخدام len(pdf_reader.pages) بدلاً من numPages
for p in range(len(pdf_reader.pages)):
    
    # الوصول للصفحة باستخدام القائمة pages
    page = pdf_reader.pages[p]
    
    # استخراج النص باستخدام extract_text()
    pdf_text.append(page.extract_text())

f.close()

# الآن يمكنك طباعة النص كاملاً أو صفحة محددة
print(f"تمت قراءة {len(pdf_text)-1} صفحات.")
print(pdf_text[1]) # لطباعة الصفحة الأولى
```

```python
import PyPDF2

# فتح الملف بوضع القراءة الثنائية
f = open('US_Declaration.pdf', 'rb')

# استخدام المسمى الحديث PdfReader
pdf_reader = PyPDF2.PdfReader(f)

# الوصول للصفحة الأولى باستخدام الفهرسة المباشرة pages[0]
first_page = pdf_reader.pages[0]

# استخراج النص باستخدام extract_text (بأحرف صغيرة وشرطة سفلية)
text = first_page.extract_text()

print(text)

# لا تنسَ إغلاق الملف
f.close()
```

```python
import PyPDF2

# 1. افتح الملف الأصلي
with open('US_Declaration.pdf', 'rb') as f_in:
    pdf_reader = PyPDF2.PdfReader(f_in)
    first_page = pdf_reader.pages[0]
    
    # 2. أنشئ الكاتب
    pdf_writer = PyPDF2.PdfWriter()
    
    # 3. أضف الصفحة (يجب أن يكون الملف f_in مفتوحاً هنا)
    pdf_writer.add_page(first_page)
    
    # 4. احفظ الملف الجديد
    with open('New_Output_File.pdf', 'wb') as f_out:
        pdf_writer.write(f_out)
```

```python
from PyPDF2 import PdfFileMerger, PdfFileReader,  PdfFileWriter

filename1 = 'F1.pdf'
filename2 = 'F2.pdf'

merger = PdfFileMerger()

merger.append(PdfFileReader(open(filename1, 'rb')))
merger.append(PdfFileReader(open(filename2, 'rb')))

merger.write("merged file 1.pdf")
```

```python
file1 = PdfFileReader(open(filename1, "rb"))
file2 = PdfFileReader(open(filename2, "rb"))

output = PdfFileWriter()

output.addPage(file1.getPage(5))
output.addPage(file2.getPage(3))


outputStream = open("merged file 2.pdf", "wb")
output.write(outputStream)
outputStream.close()
```
